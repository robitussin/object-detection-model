```python
import os

# --- CONFIGURATION ---
annotation_folder = "dataset/labels2/labels"  # Replace with your folder path

# --- Filter Function ---
def filter_yolo_annotations(file_path):
    with open(file_path, "r") as f:
        lines = f.readlines()

    # Keep only lines starting with 15 or 16
    filtered = [line for line in lines if line.strip().startswith("15") or line.strip().startswith("16")]

    # Overwrite the file with filtered lines
    with open(file_path, "w") as f:
        f.writelines(filtered)

# --- Iterate Over All TXT Files ---
for filename in os.listdir(annotation_folder):
    if filename.endswith(".txt"):
        file_path = os.path.join(annotation_folder, filename)
        filter_yolo_annotations(file_path)
        print(f"Filtered: {filename}")
```
